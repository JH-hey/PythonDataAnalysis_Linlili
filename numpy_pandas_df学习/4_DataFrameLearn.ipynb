# 第三个部分

```python
students=pd.DataFrame({
    "001":{"姓名":"小陈","考试1":85,"考试2":95,"考试3" :92},
    "002":{"姓名":"小李","考试1":91,"考试2":92,"考试3" :94},
    "003":{"姓名":"小王","考试1":86,"考试2":81,"考试3" :89},
    "004":{"姓名":"小张","考试1":79,"考试2":89,"考试3" :95},
    "005":{"姓名":"小赵","考试1":96,"考试2":91,"考试3" :91},
    "006":{"姓名":"小周","考试1":81,"考试2":89,"考试3" :92},
})
students=students.T
students
```

```python
students["考试3"]
```

```python
mean=students.loc[:,"考试1":"考试3"].mean(axis=1)
name=students["姓名"]
student_average=pd.DataFrame({"姓名":name,"平均分":mean})
student_average
```

```python
students.loc[:,"考试1":"考试3"].apply(lambda x: np.sort(x)[-2] )
```

```python
def get_grade(score):
    if score>=95:
        return "A+"
    elif score >=90:
        return "A"
    elif score >=85:
        return "B+"
    elif score >=80:
        return "B"
    elif score >=75:
        return "C+"
    else:
        return "C"
```

```python
students.loc[:,"考试1":"考试3"].map(get_grade)
```

```python
#由于dataframe中的这些数字是object类型，所以python认不出是不是数字，所以先转换
students["考试3"]=students["考试3"].astype('int')
students["考试2"]=students["考试2"].astype('int')
students["考试1"]=students["考试1"].astype('int')
students.describe()
```
