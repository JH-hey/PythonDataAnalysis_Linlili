# 双变量的图

```python
sns.scatterplot(data=test_data,x="Quantity", y="UnitPrice")
plt.show()
```

## 折线图

```python
sns.lineplot(x=test_data["Quantity"],y=test_data["UnitPrice"])
plt.show()
```

## 条形图

```python
sns.barplot(x=test_data["Country"],y=test_data["UnitPrice"])
plt.show()
#这种方法默认y轴是每个分类下的平均值
```

```python
 sns.barplot(x=test_data["Country"],y=test_data["UnitPrice"],estimator=np.max)
plt.show()
#这种方法默认y轴是每个分类下的最大值
#想要y轴是表示数量，得用计数图
```

# 三变量图

## 三变量散点图

```python
sns.scatterplot(x=test_data["UnitPrice"],y=test_data["CustomerID"],hue=test_data["Country"],size=test_data["Quantity"])
plt.legend(bbox_to_anchor=(1,1))
plt.show()
```

## 三变量条形图

```python
sns.lineplot(x=test_data["UnitPrice"],y=test_data["CustomerID"],hue=test_data["Country"])
plt.legend(bbox_to_anchor=(1,1))
plt.show()
```

## 三变量条形图

```python
sns.barplot(x=test_data["UnitPrice"],y=test_data["CustomerID"],hue=test_data["Country"])
plt.legend(bbox_to_anchor=(1,1))
plt.show()
```
