## 接下来进行假设检验：
我们采用其中某两组的数据进行两样本独立t检验，检验种类所代表的总体，关于四类数据是否具有显著差异

### 分析萼片长度

setosa与versicolor的萼片长度分布如下

```python
sns.histplot(iris_setosa['sepal length (cm)'],binwidth = 0.1)
sns.histplot(iris_versicolor['sepal length (cm)'],binwidth = 0.1)
plt.show()
```

## 建立假设检验

### $H_0$ setosa与versicolor的萼片长度平均值没有显著差异

### $H_1$ setosa与versicolor的萼片长度平均值有显著差异

### 双尾检验:单尾检验的p值=双尾的/2

```python
t_stat, p_value = ttest_ind(iris_setosa["sepal length (cm)"],iris_versicolor["sepal length (cm)"])
```

```python
print(f"t值：{t_stat}")
print(f"p值：{p_value}")
```

p值显著小于0.05
拒绝原假设： setosa与versicolor的萼片长度总体上不相等

## 分析petal width (cm)

```python
sns.histplot(iris_setosa["petal width (cm)"],binwidth=0.1)
sns.histplot(iris_versicolor["petal width (cm)"],binwidth=0.1)
plt.show()
```

```python
t_stat,p_value = ttest_ind(iris_setosa["petal width (cm)"],iris_versicolor["petal width (cm)"])
```

```python
print(f"t值：{t_stat}")
print(f"p值：{p_value}")
```

p值小于0..05，显著无差异

##  Z检验 - 与t检验差不多

```python
from statsmodels.stats.weightstats import ztest
```

双尾

```python
z_stat, p_value = ztest(iris_versicolor["sepal width (cm)"],iris_virginica["sepal width (cm)"],alternative = 'two-sided')
```

单尾：第一个均值是否大于第二个均值

```python
z_stat, p_value = ztest(iris_versicolor["sepal width (cm)"],iris_virginica["sepal width (cm)"],alternative = 'larger')
```

单尾：第一个均值是否小于第二个均值

```python
z_stat, p_value = ztest(iris_versicolor["sepal width (cm)"],iris_virginica["sepal width (cm)"],alternative = 'smaller')
```
